```python
# Pandas dataframe

WSL_table = pd.DataFrame(WSL_table[0])
```

```python
# show all table colum names

WSL_table.columns
```

```python
# create a subset to focus on the WSL Top 4 teams

top4 = ['Chelsea', 'Manchester City', 'Arsenal', 'Manchester Utd']
```

```python
# Build out a subset field which only returns data for the Top 4 subset - .isin(top4)

subset_table = WSL_table[WSL_table['Squad'].isin(top4)]
```

```python
# Build out columns of interest from above subset_table to only return data for the Top 4 subset

teams = subset_table['Squad']
win = subset_table['W']
draw = subset_table['D']
loss = subset_table['L']
```

```python
# Create a stacked bar chart to show results (wins, draws, losses) for the WSL Top 4 so far this season
plt.figure(figsize = (10,6))
plt.bar(teams, win, color = 'green')
plt.bar(teams, draw, bottom = win, color = 'yellow')
plt.bar(teams, loss, bottom = win + draw, color = 'red')

plt.xlabel('Club')
plt.ylabel('Total Number of Matches Played')
plt.legend(["Win", "Draw", "Loss"])
plt.title('Results by Club 2023/24 WSL Season')

plt.xticks(rotation = 45)

plt.show()
```
