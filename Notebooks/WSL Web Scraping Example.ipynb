```python
import pandas as pd
```

```python
# example web scrap of FBref data table 

df = pd.read_html('https://fbref.com/en/comps/189/possesion/Womens-Super-League-Stats')
```

```python
for idx, table in enumerate(df):
    print('***************************')
    print(idx)
    print(table)
```

```python
df[0]
```

```python
# rename df[0] to highlight it's the current WSL league table

league_table_df = df[0]
league_table_df
```

```python
# above table can now be used for analysis
# for example, return the WSL team with the highest attendance this season

team_with_highest_attendance = league_table_df['Attendance'].idxmax()

team_name = league_table_df.loc[team_with_highest_attendance, 'Squad']
highest_attendance = league_table_df.loc[team_with_highest_attendance, 'Attendance']

print("Team with the highest attendance:", team_name)
print("Attendance:", highest_attendance)
```
