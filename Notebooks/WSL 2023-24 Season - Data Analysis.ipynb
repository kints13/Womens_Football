```python
# import the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Women's Super League 2023-24 dataset (as at 12th March 2024)
# double up \ and add.csv to bring through file

wsl_df = pd.read_csv('WSL2324.csv')
wsl_df.head()
```

```python
# getting information on the dataset

wsl_df.info()
```

```python
# getting statistics on the dataset

wsl_df.describe()
```

```python
# total number of null values - eg. two null values in the position column

wsl_df.isna().sum()
```

```python
# converting 'Minutes' column to a float type

wsl_df['Minutes'] = wsl_df['Minutes'].astype(float)
```

```python
# create minutes per match column

wsl_df['MinsPerMatch'] = (wsl_df['Minutes'] / wsl_df['Matches_played']).astype(int)

# create goals per match column

wsl_df['GoalsPerMatch'] = (wsl_df['Gls'] / wsl_df['Matches_played']).astype(float)
print(wsl_df.head())
```

```python
wsl_df.head()
```

```python
# calculate total goals so far this season

Total_Goals = wsl_df['Gls'].sum()
print(Total_Goals)
```

```python
# calculate how many goals were penalties

Total_Penalties = wsl_df['PK'].sum()
print(Total_Penalties)
```

```python
# calculate total Penalty Attempts

Total_Penalty_Attempts = wsl_df['PKatt'].sum()
print(Total_Penalty_Attempts)
```

```python
# calculate proportion of all penalties that were scored aka penalty conversion rate

Successful_Penalties = Total_Penalties / Total_Penalty_Attempts * 100
print(Successful_Penalties)
```

```python
# pie chart for proportion of penalties scored and penalties missed

plt.figure(figsize=(13, 6))
pen_not_scored = wsl_df['PKatt'].sum() - wsl_df['PK'].sum()
data = [pen_not_scored, Total_Penalties]
labels = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# calculate the percentage of all goals scored in the WSL that were penalties

percentage_penalties = Total_Penalties / Total_Goals * 100
print(percentage_penalties)
```

```python
# How many unique positions are there?

wsl_df['Pos'].unique()
```

```python
# How many players play as a 'Forward' / 'Forward' position?

wsl_df[wsl_df['Pos'] == 'FW']
```

```python
# How many different countries/nationalities are respresented in the WSL?

np.size((wsl_df['Nation'].unique()))
```

```python
# Which country has the most players?

nationality = wsl_df.groupby('Nation').size().sort_values(ascending = False) # will print in descending order
nationality.head(10).plot(kind = 'bar', figsize=(12,6), color = sns.color_palette("magma")) # return the top 10 nations where most players come from
```

```python
# Clubs with maximum players in their squads / max squad size

wsl_df['Squad'].value_counts().nlargest(5).plot(kind = 'bar', color=sns.color_palette("viridis"))
```

```python
# clubs with the least/fewest players in the squad

wsl_df['Squad'].value_counts().nsmallest(5).plot(kind = 'bar', color=sns.color_palette("viridis"))
```

```python
# Players based on four age groups
# <20, 22-25, 25-30, 30+

Under_20 = wsl_df[wsl_df['Age'] <= 20]
age20_25 = wsl_df[(wsl_df['Age'] > 20) & (wsl_df['Age'] <= 25)]
age25_30 = wsl_df[(wsl_df['Age'] > 25) & (wsl_df['Age'] <= 30)]
Over_30 = wsl_df[wsl_df['Age'] > 30]
```

```python
# creating a pie cart of WSL players by age group

x = np.array([Under_20['Player'].count(), age20_25['Player'].count(), age25_30['Player'].count(), Over_30['Player'].count()])
mylabels = ["<=20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title('WSL Players by Age Group', fontsize = 20)
plt.pie(x, labels = mylabels, autopct = "%.1f%%")
plt.show()
```

```python
# Which WSL club has the most players Under 20 years of age?

players_under_20 = wsl_df[wsl_df['Age'] < 20]
players_under_20['Squad'].value_counts().plot(kind = 'bar', color = sns.color_palette("cubehelix"))
```

```python
# who are the Under 20 players at Bristol City?

players_under_20[players_under_20["Squad"]=='Bristol City']
```

```python
# Creating a box plot to show average age of players at each WSL club

plt.figure(figsize=(12, 6))
sns.boxplot(x = 'Squad', y = 'Age', data = wsl_df)
plt.xticks(rotation = 90)

# Arsenal have the highest average age (the highest line line within the box)
# Bristol City have an outlier - see diamond shape
# Aston Villa have the biggest spread / box
```

```python
# confirming Arsenal have the highest average age and Bristol City have the youngest average age

num_player = wsl_df.groupby('Squad').size()
data = (wsl_df.groupby('Squad')['Age'].sum()) / num_player
data.sort_values(ascending = False)
```

```python
# Total assists by WSL club

Assists_by_club = pd.DataFrame(wsl_df.groupby('Squad', as_index=False)['Ast'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Squad', y='Ast', data=Assists_by_club.sort_values(by="Ast"), palette = 'Set2')
ax.set_xlabel('Squad', fontsize=10)
ax.set_ylabel('Ast', fontsize=10)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Totals Assists by Club', fontsize = 20)
```

```python
# Top 10 Assisters in the league

top_10_assists = wsl_df[['Player', 'Squad', 'Ast', 'Matches_played']].nlargest(n=10, columns = 'Ast')
print(top_10_assists)
```

```python
# Total goals by WSL club

Goals_by_club = pd.DataFrame(wsl_df.groupby('Squad', as_index=False)['Gls'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Squad', y='Gls', data=Goals_by_club.sort_values(by="Gls"), palette = 'rocket')
ax.set_xlabel('Squad', fontsize=10)
ax.set_ylabel('Gls', fontsize=10)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Totals Goals by WSL Club', fontsize = 20)
```

```python
# Top WSL goalscorer so far this season

top_10_goalscorers = wsl_df[['Player', 'Squad', 'Gls', 'Matches_played']].nlargest(n=10, columns = 'Gls')
top_10_goalscorers
```

```python
# Goals scored per match

top_10_goals_per_match = wsl_df[['Player', 'GoalsPerMatch', 'Matches_played', 'Gls']].nlargest(n=10, columns = 'GoalsPerMatch')
top_10_goals_per_match

# Khadija 'Bunny' Shaw is scoring more than a goal a game(!)
```

```python
# Pie chart - Goals with an assist and goals scored without an assist

plt.figure(figsize=(14,7))
assists = wsl_df['Ast'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# Top 10 players with most yellow cards

wsl_yellow = wsl_df.sort_values(by = 'CrdY', ascending = False)[:10]
plt.figure(figsize = (20,6))
plt.title("Players with the most Yellow Cards")
c=sns.barplot(x = wsl_yellow['Player'], y = wsl_yellow['CrdY'], label = 'Players', color = 'yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(), rotation = 45)
c
```
